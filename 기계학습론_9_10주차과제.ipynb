#데이터 전처리

```python
from collections import Counter,defaultdict

def outlier_detection(df, n, columns):
    rows = []
    will_drop_train = []
    outlier_count = defaultdict(int)
    for col in columns:
        Q1 = np.nanpercentile(df[col], 25)
        Q3 = np.nanpercentile(df[col], 75)
        IQR = Q3 - Q1
        outlier_point = 1.5 * IQR
        outlier_rows = df[(df[col] < Q1 - outlier_point)|(df[col] > Q3 + outlier_point)].index
        rows.extend(outlier_rows)
        outlier_count[col] += len(outlier_rows)
    for r, c in Counter(rows).items():
        if c >= n: will_drop_train.append(r)
    return will_drop_train,outlier_count

will_drop_train,outlier_count = outlier_detection(df, 5, df.select_dtypes(["float", "int"]).columns)
print(outlier_count)
print(len(will_drop_train))
```

```python
df.drop(will_drop_train, inplace = True, axis = 0)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit_transform(df["gender"])
df["gender"] = le.fit_transform(df["gender"])

le = LabelEncoder()
le.fit_transform(df["tartar"])
df["tartar"] = le.fit_transform(df["tartar"])

le = LabelEncoder()
le.fit_transform(df["oral"])
df["oral"] = le.fit_transform(df["oral"])
```

```python
y = df["smoking"]
X = df.drop("smoking", axis = 1)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= 0.2,
                                                 random_state = 13)
```

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler

scaler = StandardScaler()
scaler.fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

#모델 학습

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(random_state = 13)
dt_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = dt_model.predict(X_test)
accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
print(cm)
```

# 특성 선택

```python
sort = dt_model.feature_importances_.argsort()
plt.barh(df.columns[sort], dt_model.feature_importances_[sort])
plt.xlabel("Feature Importance")
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=100, random_state = 13)
rf.fit(X_train, y_train)

y_pred = rf.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

```python
sort = rf.feature_importances_.argsort()
plt.barh(df.columns[sort], rf.feature_importances_[sort])
plt.xlabel("Feature Importance")
```

```python
remove_features = ['hearing(left)', 'hearing(right)','oral','Urine protein','dental caries']
column_names = df.columns[df.columns != 'smoking'].tolist()
selected_features = list(set(column_names) - set(remove_features))

X_train_df = pd.DataFrame(X_train, columns=column_names)
X_test_df = pd.DataFrame(X_test, columns=column_names)

X_train_selected = X_train_df[selected_features]
X_test_selected = X_test_df[selected_features]
```

# 모델 재학습

```python
dt_model_selected = DecisionTreeClassifier(random_state = 13)
dt_model_selected.fit(X_train_selected, y_train)

# 예측 및 성능 평가
y_pred_selected = dt_model_selected.predict(X_test_selected)
print(accuracy_score(y_test, y_pred_selected))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred_selected))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_selected)
print(cm)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_seleccted = RandomForestClassifier(n_estimators=100, random_state = 13)
rf_seleccted.fit(X_train_selected, y_train)

y_pred_selected = rf_seleccted.predict(X_test_selected)
print(accuracy_score(y_test, y_pred_selected))
```

# 성능평가

```python
print(classification_report(y_test, y_pred_selected))
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred_selected)
print(cm)
```

```python
from sklearn.metrics import roc_curve, auc,roc_auc_score

rf_roc_auc = roc_auc_score(y_test, rf_seleccted.predict(X_test_selected))
fpr, tpr, thresholds = roc_curve(y_test, rf_seleccted.predict_proba(X_test_selected)[:,1])

plt.figure(figsize=(8,8))
plt.plot(fpr, tpr, label = 'AUC (area = %0.2f)' % rf_roc_auc)
plt.plot([0, 1], [0, 1],'g--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title("ROC.AUC :")
plt.legend()
plt.show()
```
